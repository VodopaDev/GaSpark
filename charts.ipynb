```python
import matplotlib.pyplot as plt
import numpy as np

result = open("result/resultstationcomp")
header = result.readline().split(";")
vals = [s.split(";")for s in result.readlines()]
```

```python
x = [s[0] for s in vals]
y = [(float(s[1])/1000, float(s[2])/1000) for s in vals]

figure = plt.figure(figsize=(16,9))
xy = plt.plot(x,y, figure=figure)
plt.title("Highway and road gas stations price comparison\n", size=30)
plt.legend(["Highway","Road"], loc=0, fontsize=20)
plt.xlabel("\nDate", size=20)
plt.ylabel("Price (€/L)\n", size=20)
```

```python
x = [s[0] for s in vals]
y = [(float(s[1])/float(s[2]) - 1) * 100 for s in vals]

figure = plt.figure(figsize=(16,9))
xy = plt.plot(x,y, figure=figure)
plt.title("Highway vs road gas stations percent difference\n", size=30)
plt.xlabel("\nDate", size=20)
plt.ylabel("Price difference (in %)\n", size=20)
```

```python
x = [s[0] for s in vals]
y = [(float(s[1])-float(s[2]))/10 for s in vals]

figure = plt.figure(figsize=(16,9))
xy = plt.plot(x,y, figure=figure)
plt.title("Highway vs road gas stations gross difference\n", size=30)
plt.xlabel("\nDate", size=20)
plt.ylabel("Price difference (in c/L)\n", size=20)
```
